```python
import numpy as np
import matplotlib.pyplot as pt
import math
```

```python
def ReLU(x):
    return max(0,x)

def derivativeReLU(x):
    return x>0

def sigmoid(x):
    r = 1/(1+ math.exp(-x))
    return r
```

```python
a0 = 0.25
a1 = 0.15
b0 = 0
y = 0.2345

eta = 0.1
epoch = 1000

w0 = np.random.rand(1)
w1 = np.random.rand(1)

print(str(w0) + " "+str(w1))


for i in range(0,epoch):

 # forward pass

    z = w0*a0 + w1*a1 + b0

     # Activation function
    o = sigmoid(z)
     #sigmoid

     #cost
    cost = (o-y)**2

    #backProp
    

    db0 = (o-y)*derivativeReLU(o)            #db = rate of change of C wrt b0
    dw0 = (o-y)*derivativeReLU(o)*a0         #dw0
    dw1 = (o-y)*derivativeReLU(o)*a1         #dw1

    #update the parameters
    b0 = b0 - eta*db0
    w0 = w0 - eta*dw0
    w1 = w1 - eta*dw1
    print("output = "+str(o) + " Cost= "+str(cost))
    pt.scatter(o,cost)
   
```
